# Part1: Working with XML

```python
Below, I am going to perform XML Parsing of Website.
```

```python
# import class from module xmlparser
from xmlparser import SitemapParser
# Give url
website_url = "https://www.epicurious.com"
# create the object for class SitemapParser
sitemap_parser = SitemapParser(website_url)  
# Call the method which returns text format of the "website_url/robotstxt"
sitemap_parser.fetch_robots_txt()
```

Result : Got the all sitemaps along with allow and Disallow results

```python
# method extract_sitemaps call to get exactly how many sitemaps contains a certain website_url (using object call).
sitemap_parser.extract_sitemaps()
```

Result : Here, we can clearly see, given website contains 3 sitemaps

```python
# converted xml of all 3 sitemaps into dataframe
sitemap_parser.run()
```

Result : Here I get the dataframe from sitemaps of website.

conclusion : I have created module(xmlparser.py) and in that class called "SitemapParser" created to perform several task by creating many methods.

# Part - 2: Web page Scraping Using an API 

Below, I am going perform Web page Scraping using API.

```python
# Reading the API key
api_key = 'pub_331919af4c2a25bb759ef095c96b1f3706e05'

# Reading the API from the website
base_url = f'https://newsdata.io/api/1/news?apikey={api_key}&q=education&country=us&language=en&category=business'

from webapi import APIDataFetch as adf

# object created for class
obj_adf = adf(api_key,base_url)
# call the method to get data from website by using make_request() method
obj_adf.make_request()
```

Result : Correctly fetched the data from website

```python
# method call using object, to get the data as dataframe
data = obj_adf.fetch_data()
data
```

Result : Scrapped the data and transformed into the  dataframe.

# Data Cleaning

The Data Frame contians a lot of columns which might not be very useful for the analysis. We are here dropping the coloumns that tells very less about the information we need.

```python
from datacleaning import DataCleaning
obj_datacleaning = DataCleaning(data)
data = obj_datacleaning.feild_drop()
data
```

Result : Dropped the feild which is not goimg to help in Data analysis
